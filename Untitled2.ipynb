```python
import numpy as np
m1 = np.random.randint(1,20,15)
print("Array")
print(m1)
# Reshape the array to 3 by 5
m2=np.reshape(m1,(3,5))
print("Reshaped Array")
print(m2)
#Printing array shape.
print("Array Shape")
print(m2.shape)
#Replace the max in each row by 0
index = m2.argmax(axis=1)
print("indexes of max values in array")
print(index)
print("Array with replaced max value")
m2[0][index[0]] = 0
m2[1][index[1]] = 0
m2[2][index[2]] = 0
print(m2)

```

```python
# import numpy library
import numpy as np

# creating numpy 2d-array
m = np.array([[3,-2],[1,0]])

print("Printing the Original square array:\n",m)

# find eigenvalues and eigenvector
w, v = np.linalg.eig(m)

# print eigen values
print("Printing the Eigen values of the given square array:\n",w)

# print eigen vectors
print("Printing Right eigenvectors of the given square array:\n",v)


```
